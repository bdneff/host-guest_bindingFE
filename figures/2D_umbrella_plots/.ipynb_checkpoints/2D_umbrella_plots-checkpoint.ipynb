```python
import numpy as np
import os
from matplotlib import pyplot
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize,  ListedColormap
from scipy.interpolate import griddata
import numpy as np
from contextlib import redirect_stdout
from matplotlib import colors
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import shutil
import os
import glob
import copy
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.cm as cm
from matplotlib.patches import Rectangle
import math

# Function to find distance
def shortest_distance(x1, y1, z1, a, b, c, d):
     
    dist = abs((a * x1 + b * y1 + c * z1 + d))
    e = (math.sqrt(a * a + b * b + c * c + d * d))
    #print("Perpendicular distance is", dist/e)
    return dist/e

def read_wham_outfiles(file_path):
    # Replace 'filename.txt' with your actual file path
    #file_path =  "/Users/bdneff/Desktop/FE-profile/GAFF/DRG/TEST/all_pull/outfile.out"

    # Load the file as a 2D array of data
    data = np.loadtxt(file_path, usecols=(0, 1, 2, 3))
    x = data[:1]
    y = data[:2]
    fe = data[:3]
    prob = data[:4]
    
    return data
    

class gro_data:
    def __init__(self, path_to_gro):
        self.path_to_gro = path_to_gro
        #self.new_path_to_gro = path_to_gro + '.new'
        self.newlist = None
        self.time = None
        self.distances = None
        self.distances_rot = None
        self.dirs = None
        self.orth_vec = None
        
        
    def gro_to_list(self):
        with open(self.path_to_gro) as f:
            lines = f.readlines() 
        self.newlist = [line.split() for line in lines]
        f.close()
        return self.newlist
        
    def get_frame_count(self):
        frame_count = 0
        for line in self.newlist:
            if isinstance(line, str) and line.startswith("GRoups of Organic Molecules in ACtion for Science in water t="):
                frame_count += 1
        return frame_count


    #def rotate_and_update_coordinates(self):
    # Perform the rotation (similar to the previous implementation)
    # This method should update self.newlist with rotated coordinates
    # ...
        
    def measure_cavity_displacement(self):
        p1 = []
        p2 = []
        p3 = []
        pN = []
        for line in self.newlist:
            if len(line) > 2:
                if line[1] == 'C36':
                    p1.append(np.array([float(line[3]),float(line[4]),float(line[5])]))
                if line[1] == 'C37':
                    p2.append(np.array([float(line[3]),float(line[4]),float(line[5])]))
                if line[1] == 'C39':
                    p3.append(np.array([float(line[3]),float(line[4]),float(line[5])]))
                if line[0] == '2LIG' and line[1] == 'CA':
                    pN.append(np.array([float(line[3]),float(line[4]),float(line[5])]))

        self.distances = []
        self.time = []
        for i in range(len(p1)):
            self.time.append(2 * i)
            # These two vectors are in the plane
            v1 = p3[i] - p1[i]
            v2 = p2[i] - p1[i]

            # the cross product is a vector normal to the plane
            cp = np.cross(v1, v2)
            #print(cp)
            a, b, c = cp
            self.orth_vec = cp
            # This evaluates a * x3 + b * y3 + c * z3 which equals d
            d = np.dot(cp, p3[i])

            x1 = pN[i][0]
            y1 = pN[i][1]
            z1 = pN[i][2]
        
            #print('The equation is {0}x + {1}y + {2}z + {3} = 0'.format(a, b, c, d))
            val = (shortest_distance(x1, y1, z1, a, b, c, d))
            self.distances.append(val)

    
        #self.distances = []
        #self.time = []
        #self.distances_rot = []
        #for i in range(len(p1)):
            #self.time.append(2 * i)
            #v1 = p3[i] - p1[i]
            #v2 = p2[i] - p1[i]

            #normal = np.cross(v1, v2)
            #normal = normal / np.linalg.norm(normal)  # Normalize the vector

            # Align this normal with the Z-axis
            #axis_to_align = np.array([0, 0, 1])
            #rotation_matrix = self.rotation_matrix_from_vectors(normal, axis_to_align)

            # Apply this rotation to all coordinates
            #rotated_pN = np.dot(rotation_matrix, pN[i])
            #self.distances_rot.append(rotated_pN)
            #self.newlist=rotated_pN
                
            #with open(self.new_path_to_gro, 'w') as f:
            #    for line in self.newlist:
            #        f.write(' '.join(map(str, line)) + '\n')
        return self.distances
    
    
    @staticmethod
    def rotation_matrix_from_vectors(vec1, vec2):
        """ Find the rotation matrix that aligns vec1 to vec2 """
        a, b = (vec1 / np.linalg.norm(vec1)).reshape(3), (vec2 / np.linalg.norm(vec2)).reshape(3)
        v = np.cross(a, b)
        c = np.dot(a, b)
        s = np.linalg.norm(v)
        kmat = np.array([[0, -v[2], v[1]], [v[2], 0, -v[0]], [-v[1], v[0], 0]])
        rotation_matrix = np.eye(3) + kmat + kmat.dot(kmat) * ((1 - c) / (s ** 2))
        return rotation_matrix
            
    def measure_orientation(self):
        f1 = []
        fN = []
        g1 = []
        gN = []
        for line in self.newlist:
            if len(line) > 2:
                if line[1] == 'C34':
                    f1.append(np.array([float(line[3]),float(line[4]),float(line[5])]))
                if line[0] == '2LIG' and line[1] == 'OT2':
                    fN.append(np.array([float(line[3]),float(line[4]),float(line[5])]))
                if line[1] == 'C35':
                    g1.append(np.array([float(line[3]),float(line[4]),float(line[5])]))
                if line[0] == '2LIG' and line[1] == 'N':
                    gN.append(np.array([float(line[3]),float(line[4]),float(line[5])]))
        self.distances_rot = []
        for i in range(len(self.distances)):
            #self.time.append((2 * i)/1000)
            # These two vectors are in the plane
            v1 = f1[i] - fN[i]
            v2 = g1[i] - gN[i]
            v3 = f1[i] - gN[i]
            v4 = g1[i] - fN[i]
            rot_dist = (np.linalg.norm(v1) + np.linalg.norm(v2)) - (np.linalg.norm(v3)+np.linalg.norm(v4))
            self.distances_rot.append(rot_dist)
        
    
    def check_flip(self):
        n1 = []
        n2 = []
        for line in self.newlist:
            if len(line) > 2:
                if line[0] == '2LIG' and (line[1] == 'NZ' or line[1] == 'CZ'):
                    n1.append(np.array([float(line[3]),float(line[4]),float(line[5])]))
                if line[0] == '2LIG' and line[1] == 'CA':
                    n2.append(np.array([float(line[3]),float(line[4]),float(line[5])]))
        self.dirs = []
        for i in range(len(self.distances)):
            vec_dir = n1[i] - n2[i]
            start_vec = n1[0] - n2[0]
            self.dirs.append(np.dot(vec_dir,start_vec)/np.dot(start_vec,start_vec))
        return self.dirs

    def plot_distance(self):
        fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(25,7))
        plt.subplot(1, 2, 1)
        plt.plot(self.time,self.distances)
        plt.xlabel('time (ps)')
        plt.ylabel('min perpendicular distance (nm)')
        plt.ylim(0,3)
        
        plt.subplot(1, 2, 2)
        plt.hist(self.distances, bins=50)
        plt.ylabel('occurences')
        plt.xlabel('distances')

    def plot_orientation(self):
        fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(25,7))
        plt.subplot(1, 2, 1)
        plt.plot(self.time,self.distances_rot)
        plt.xlabel('time (ns)')
        plt.ylabel('orientation of amino acid')
        plt.ylim(-.7,.7)
        
        plt.subplot(1, 2, 2)
        plt.hist(self.distances_rot, bins=80)
        #plt.ylabel('occurences')
        #plt.xlabel('distances')

    def plot_2d_hist(self, ax):
        #ax.set_ylabel('perpendicular distance')
        #ax.set_xlabel('orientation of ligand')
        ax.hist2d(self.dirs, self.distances, bins=(40, 40), cmap='magma_r', range=[[-1, 1], [0, 5]])
        # ax.set_ylim(0, 10)  # Optionally set limits for y-axis
        # ax.set_xlim(-0.5, 0.5)  # Optionally set limits for x-axis
        # ax.set_title('Your Title')  # Optionally set a title
        # plt.show()  # Don't call plt.show() here
        
    def plot_lone_2d_hist(self):
        fig, ax = plt.subplots()
        ax.set_ylabel('perpendicular distance')
        ax.set_xlabel('orientation of ligand')
        ax.hist2d(self.dirs, self.distances, bins=(40, 40), cmap='magma_r', range=[[-1, 1], [0, 5]])
        #ax.set_ylim(0, 10)  # Optionally set limits for y-axis
        #ax.set_xlim(-0.5, 0.5)  # Optionally set limits for x-axis
        #ax.set_title('Your Title')  # Optionally set a title
        #plt.show()  # Don't call plt.show() here
        
def process_gro_file(gro_file_path):
    # Load data once and store it in memory if possible
    gro = gro_data(gro_file_path)
    gro.gro_to_list()
    falls = gro.measure_cavity_displacement()
    flaps = gro.check_flip()
    
    flip_count = len([1 for i in range(1, len(flaps)) if flaps[i-1] * flaps[i] < 0.0 ])

    resolution = 20  # Constant resolution value
    dat_length = 100000
    gro_frame_count = 5000

    # Calculate frames to omit based on the formula
    frames_to_omit = int(gro_frame_count - (dat_length * (gro_frame_count / resolution)))

    # Omit the specified number of frames from the beginning
    flaps = flaps
    falls = falls
    
    
    flips = [x for x, ele in enumerate(flaps) if ele < 0]  # Adjust the threshold as needed
    fallouts = [y for y, ele in enumerate(falls) if ele > 3.0]  # Adjust the threshold as needed

    # Count the number of gaps in the flips list

    return flips, fallouts, flip_count

# Function to remove frames from DAT file
def remove_frames_from_dat(dat_file_path, frames_to_remove, new_subdir, end_point=100000):
    parent_directory = os.path.join(os.path.dirname(dat_file_path), 'processed')
    # Create the new_subdir directory if it doesn't exist
    new_subdir_path = os.path.join(parent_directory, new_subdir)
    os.makedirs(new_subdir_path, exist_ok=True)
    
    with open(dat_file_path, 'r') as dat_file:
        lines = dat_file.readlines()
        lines = lines[:end_point]
        
    for i in sorted(frames_to_remove, reverse=True):
        if i < len(lines):
            del lines[i]

    new_dat_file_path = os.path.join(new_subdir_path, os.path.basename(dat_file_path))
    
    with open(new_dat_file_path, 'w') as new_dat_file:
        new_dat_file.writelines(lines)
        
def process_files_in_directory(gro_directory, dat_directory, processed_dat_directory):
    # Create the 'processed' directory if it doesn't exist
    os.makedirs(processed_dat_directory, exist_ok=True)
    
    source_metadat_file = os.path.join(dat_directory, 'metadat.dat')
    destination_metadat_file = os.path.join(processed_dat_directory, 'metadat.dat')
    shutil.copy(source_metadat_file, destination_metadat_file)
    
    
    # Get a list of GRO files in the 'filter' directory
    gro_files = [f for f in os.listdir(gro_directory) if f.endswith('.gro')]
    metadat_file_path = os.path.join(processed_dat_directory, 'metadat.dat')
    lines_to_remove = []
    for i, gro_file in enumerate(gro_files):
        gro_file_path = os.path.join(gro_directory, gro_file)
        #print(gro_file_path)
        flips, fallouts, flip_count = process_gro_file(gro_file_path)
        #print(flip_count)
        #print(len(flips))
        #print(flip_count)
        
        with open(metadat_file_path, 'r') as metadat_file:

            if (len(flips) > 0 and flip_count < 10) or len(fallouts) >= 1:
                #print('blah')
                # Read the metadat.dat file
                metadat_lines = metadat_file.readlines()

                # Determine the corresponding line to remove based on the DAT file name
                #gro_file_name = os.path.splitext(gro_file)[0]
                dat_file_name = f'us_{i}_pullx.dat'
                line_to_remove = None
                for line_index, line in enumerate(metadat_lines):
                    if dat_file_name in line:
                        line_to_remove = line_index
                        just_us_name = metadat_lines[line_index].split()
                        lines_to_remove.append(just_us_name[0])
                        break
                #print(len(metadat_lines))
                # Remove the line corresponding to the DAT file name

                #if line_to_remove is not None:
                    #metadat_lines.pop(line_to_remove)
                # Write the modified metadat.dat content back to the file

                #with open(metadat_file_path, 'w') as metadat_file:
                    #metadat_file.writelines(metadat_lines)

        # Determine the corresponding DAT file based on the GRO file name
        gro_file_name = os.path.splitext(gro_file)[0]
        dat_file_name = f'us_{i}_pullx.dat'
        dat_file_path = os.path.join(dat_directory, dat_file_name)

        # Adjust frame indices for DAT file (assuming 20x lower resolution)
        adjusted_frames_to_remove = []
        for frame_index in flips:
            adjusted_frames_to_remove.extend(range(frame_index * 20, (frame_index + 1) * 20))
        #print(adjusted_frames_to_remove)

        # Remove frames from DAT file
        for i in range(0,50):
            omit_ends = 100000 - (i*2000)
            remove_frames_from_dat(dat_file_path, adjusted_frames_to_remove, f'data{i}', omit_ends)
    #print(lines_to_remove)
    
    with open(metadat_file_path, 'r') as file:
        lines = file.readlines()

    # Keep lines that don't start with any of the specified strings
    filtered_lines = [line for line in lines if not any(line.startswith(s) for s in lines_to_remove)]

    # Write the filtered lines back to the file or a new file
    with open(metadat_file_path, 'w') as file:  # Use 'w' to overwrite the same file
        file.writelines(filtered_lines)

    for i in range(0,50):
        source_metadat_file = os.path.join(processed_dat_directory, 'metadat.dat')
        destination_metadat_file = os.path.join(processed_dat_directory, f'data{i}/metadat.dat')
        shutil.copy(source_metadat_file, destination_metadat_file)
        
def process_unbound(dat_directory, processed_dat_directory):
    # Create the 'processed' directory if it doesn't exist
    os.makedirs(processed_dat_directory, exist_ok=True)

    # Copy the metadat file
    source_metadat_file = os.path.join(dat_directory, 'metadat.dat')
    
    # Find all us_*.dat files
    us_files = [f for f in os.listdir(dat_directory) if f.startswith('us_') and f.endswith('.dat')]

    for i in range(50):
        # Create subdirectory for each data set
        subdirectory = os.path.join(processed_dat_directory, f'data{i}')
        os.makedirs(subdirectory, exist_ok=True)

        # Copy metadat file to each subdirectory
        destination_metadat_file = os.path.join(subdirectory, 'metadat.dat')
        shutil.copy(source_metadat_file, destination_metadat_file)

        # Process each us_*.dat file
        for us_file in us_files:
            source_file = os.path.join(dat_directory, us_file)
            destination_file = os.path.join(subdirectory, us_file)

            # Copy and truncate the us_*.dat files
            with open(source_file, 'r') as source, open(destination_file, 'w') as destination:
                lines = source.readlines()
                if i>0:
                    destination.writelines(lines[:-i * 2000]) # Remove last i * 10000 lines
                else:
                    destination.writelines(lines)

# Create the 'jet' colormap
jet = plt.cm.get_cmap('jet', 512)

# Get the colors of the 'jet' colormap
colors = jet(np.linspace(0, 1, 512))
colors = colors[:480]

# Determine the maximum color of the 'jet' colormap (for the upper half uniform color)
max_color = colors[-1]

# Change the second half of the colormap to the maximum color
for i in range(len(colors)):
    if i%2 == 0:
        colors[int(i/2)] = colors[i]
    
colors[240:] = max_color
bottom_half_colormap = ListedColormap(colors[:240])

# Create a new colormap from the modified color array
custom_jet = LinearSegmentedColormap.from_list('UpperHalfMaxColor', colors)

class wham_output:
    def __init__(self, path_to_out):
        self.path_to_out = path_to_out
        #self.x = None
        #self.y = None
        #self.fe = None
        #self.prob = None
        outdata = read_wham_outfiles(self.path_to_out)
        self.x = outdata[:,0]
        self.y = outdata[:,1]
        self.fe = outdata[:,2]
        self.prob = outdata[:,3]
        
        if len(self.x) != len(self.y):
            print('this is not a square FE surface')
        else:
            self.bins = int(np.sqrt(len(self.x)))
            self.dphi = 2 * 3.14 / self.bins 
            
            self.Prob = [np.exp(-(ele * 1000) / (8.3145 * 300)) * 0.157**2 for ele in outdata[:, 2]]
            self.norm = np.sum(self.Prob)
    
    def print_bin_info(self):
         print("there are " + str(self.bins) + " bins in x and y and the bin size is " + str(self.dphi) + " radians")
    
    def calculate_biased_prob(self, og_phi1, og_phi2, k1, k2 ):
        self.biased_prob = []
        phi1 = og_phi1 * np.pi/180
        phi2 = og_phi2 * np.pi/180
        for i in range(len(self.fe)):
            cont1 = np.exp(-(0.5 * k1 * (self.x[i] - phi1)**2) / (8.3145* 0.001 * 300))
            cont2 = np.exp(-(0.5 * k2 * (self.y[i] - phi2)**2) / (8.3145 * 0.001 * 300))
            cont3 = np.exp(-(self.fe[i] * 1000) / (8.3145 * 300))
            val = 1/self.norm * cont1  * cont2 * cont3 
            #print(1/self.norm)
            self.biased_prob.append(val)
            #print(val)
        self.biased_prob_integral = np.sum(np.multiply(self.biased_prob,self.dphi**2))
        #print(self.biased_prob_integral)
        print('QOverQPrime = ' + str(1/self.biased_prob_integral) )
        return 1/self.biased_prob_integral
    
    def contour_plot(self, ax, z_attr_name, cmap='jet', cbar_label='$\Delta$G kJ/mol', cbar_orientation='vertical', vmin=None, vmax=None, colorbar=True):
        if ax == 0:
            fig, ax = plt.subplots()
        
        x = self.x * 180/np.pi
        y = self.y * 180/np.pi
        z = getattr(self, z_attr_name)  # Dynamically get the z attribute

        # Create grid coordinates
        xi = np.linspace(min(x), max(x), 40)
        yi = np.linspace(min(y), max(y), 40)
        xi, yi = np.meshgrid(xi, yi)

        # Interpolate z values on grid coordinates
        zi = griddata((x, y), z, (xi, yi), method='linear')
        
        # Define the levels of the contour if vmin and vmax are provided
        if vmin is not None and vmax is not None:
            levels = np.linspace(vmin, vmax, 1000)
        else:
            levels = 1000

        # Create contour plot on the provided axes
        chose_cmap = plt.cm.get_cmap(cmap)

        # Make a copy of the colormap to avoid changing the original colormap
        #chose_cmap = chose_cmap.copy()

        # Set the 'over' color to gray
        
        chose_cmap.set_over('darkgray')

        # Create contour plot on the provided axes
        contour = ax.contourf(xi, yi, zi, levels=levels, cmap=chose_cmap, vmin=vmin, vmax=vmax, extend='max')

        if colorbar == True:
            # Create colorbar and add label
            divider = make_axes_locatable(ax)
            cax = divider.append_axes("right", size="5%", pad=0.2) if cbar_orientation == 'vertical' else divider.append_axes("bottom", size="5%", pad=0.5)

            # Only create ScalarMappable with vmin and vmax if they are provided
            if vmin is not None and vmax is not None:
                norm = Normalize(vmin=vmin, vmax=vmax)
            else:
                norm = None

            # Create the ScalarMappable and colorbar
            sm = plt.cm.ScalarMappable(cmap=cmap, norm=Normalize(vmin=vmin, vmax=vmax))
            sm.set_array([])  # You must set_array for ScalarMappable
            cbar = plt.colorbar(sm, cax=cax, orientation=cbar_orientation)
            if cbar_label =='$\Delta$G kJ/Mol':
                labelpad = 30
            else:
                labelpad = 60
            cbar.set_label(cbar_label, rotation=270, labelpad=labelpad, fontsize=30)
            cbar.ax.tick_params(labelsize=16)

            # Set plot labels and title
            ax.set_xlabel('$\Phi_1$ (degrees)', fontsize=32)
            ax.set_ylabel('$\Phi_2$ (degrees)', fontsize=32)
            ax.tick_params(labelsize=24)
            ax.set_xticks([-90, 0, 90])
            ax.set_yticks([-90, 0, 90])
        
        return zi
    
    def colored_scatter(self, ax, z_attr_name, cmap='jet', cbar_label='$\Delta$G kJ/mol', cbar_orientation='vertical', vmin=None, vmax=None):
        if ax == 0:
            fig, ax = plt.subplots()
        
        x = self.x * 180/np.pi
        y = self.y * 180/np.pi
        z = getattr(self, z_attr_name)  # Dynamically get the z attribute

        # Create the scatter plot
        # Create contour plot on the provided axes
        chose_cmap = plt.cm.get_cmap(cmap)

        # Make a copy of the colormap to avoid changing the original colormap
        #chose_cmap = chose_cmap.copy()

        chose_cmap.set_over('darkgray')
        scatter = ax.scatter(x, y, c=z, cmap=chose_cmap, vmin=vmin, vmax=vmax, marker = 's', s = 50)

        # Create colorbar and add label
        #divider = make_axes_locatable(ax)
        #cax = divider.append_axes("right", size="5%", pad=0.2) if cbar_orientation == 'vertical' else divider.append_axes("bottom", size="5%", pad=0.5)
        
        # Create the ScalarMappable and colorbar
        #sm = plt.cm.ScalarMappable(cmap=cmap, norm=Normalize(vmin=vmin, vmax=vmax))
        #sm.set_array(z)  # Set array for ScalarMappable
        #cbar = plt.colorbar(sm, cax=cax, orientation=cbar_orientation)
        #if cbar_label =='$\Delta$G kJ/Mol':
        #    labelpad = 30
        #else:
        #    labelpad = 60
        #cbar.set_label(cbar_label, rotation=270, labelpad=labelpad, fontsize=30)
        #cbar.ax.tick_params(labelsize=16)
        
        # Set plot labels and title
        ax.set_xlabel('$\Phi_1$ (degrees)', fontsize=32)
        ax.set_ylabel('$\Phi_2$ (degrees)', fontsize=32)
        ax.tick_params(axis='both', which='major', length=10, width=2) 
        ax.tick_params(labelsize=24)
        ax.set_xticks([-90, 0, 90])
        ax.set_yticks([-90, 0, 90])
        ax.set_ylim(-180,180)
        ax.set_xlim(-180,180)
        
        # Remove borders from the plot
        #ax.spines['top'].set_visible(False)
        #ax.spines['right'].set_visible(False)
        #ax.spines['left'].set_visible(False)
        #ax.spines['bottom'].set_visible(False)
        
```
